## 2. The question: decision, action, cost of a wrong call

- **What decision does this improve?**
  Deciding which specific pages (content assets) in a client's portfolio should be assigned to an editor for refresh review first.
- **Who acts on the output, and what do they do?**
  A content editor or content manager acts on the ranked list. They review the page's search console and analytics metrics alongside its reason codes (e.g., `thin_visible_page`, `page_one_decay_risk`) and determine the specific update type (e.g., rewrite title/meta, add content depth, or monitor it).
- **What does a wrong answer cost?**
  - *False Positive (recommending a healthy or stable page for refresh)*: Wasted writing/editorial hours (often valued at $100–$300 per article) on content that did not need updates.
  - *False Negative (missing a page that is in severe decay)*: Compounded loss of search ranking position, organic traffic, and conversion revenue over time.
- **Why does data or ML help at all?**
  A simple rule (e.g., "refresh every page older than 180 days") is inefficient because content decay is non-linear and multidimensional. Some old pages remain highly stable, while some newer pages decay rapidly due to search keyword volume shifts. ML allows us to weight and combine multiple signals (impressions, position drift, CTR, and scroll/engagement rates) to capture complex decay patterns that simple manual thresholds miss.

## 3. Quick look at the data (2-3 real numbers)

To confirm that Lane 2 is a high-leverage direction, we computed key metrics from the starter dataset:
1. **Decline Prevalence**: Out of 30,000 pages, **16,262 (54.2%)** are currently in a declining trend.
2. **High-Visibility Exposure**: **9,961** of these declining pages (61.3%) still maintain high search console visibility (`impressions_90d >= 500`). These represent high-leverage refresh candidates where traffic recovery is highly valuable and realistic.
3. **Age is Not a Simple Proxy**: The median age of declining pages is **216 days**, whereas non-declining pages have a median age of **287 days**. This counterintuitive result demonstrates that older content is not automatically the most prone to decay. A simple age-based cutoff rule would miss key declining assets and waste budget on stable older pages.

```python
# Load starter data and extract key metrics
import pandas as pd
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

total_pages = len(df)
declining_mask = df["trend_direction"].str.lower().eq("down")
declining_pages = declining_mask.sum()
high_vis_declining = ((declining_mask) & (df["impressions_90d"] >= 500)).sum()
median_age_declining = df[declining_mask]["content_age_days"].median()
median_age_stable = df[~declining_mask]["content_age_days"].median()

print(f"1. Total Pages: {total_pages:,}")
print(f"2. Declining Pages: {declining_pages:,} ({declining_pages/total_pages:.1%})")
print(f"3. High-Visibility Declining Pages (impressions >= 500): {high_vis_declining:,} ({high_vis_declining/declining_pages:.1%} of declining)")
print(f"4. Median Age (Declining): {median_age_declining:.1f} days")
print(f"5. Median Age (Stable/Up): {median_age_stable:.1f} days")
```

## 4. Careful words: what I can and can't claim

- **What I CAN claim**:
  - We can claim **observed associations** between historical search/engagement features (e.g., impressions, CTR, position, age) and page decline.
  - We can claim **decision-support** utility, helping editors prioritize their limited editing capacity by ranking pages by their historical probability of decline.
  - We can claim **directional** findings showing where traffic and ranking indicators suggest decay.
- **What I CANNOT claim**:
  - We cannot claim **causal proof**. We cannot state that updating a page *causes* a recovery, as that would require a randomized control experiment or causal inference modeling.
  - We cannot claim to have "predicted Google's algorithm" or identified official search engine ranking factors. Our analysis is strictly descriptive and client-specific search console telemetry.

```python
# Demonstrate limit of simple linear correlations
import pandas as pd
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")
y = df["trend_direction"].str.lower().eq("down").astype(int)

print("Correlation with decline label:")
print(f" - content_age_days: {df['content_age_days'].corr(y):.4f}")
print(f" - ctr: {df['ctr'].corr(y):.4f}")
print("The weak correlations indicate that a simple linear claim about single variables is insufficient.")
```
